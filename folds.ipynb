```python
import pandas as pd
from src.evt_ppo.data import load_local, prices_to_log_returns
from src.evt_ppo.walkforward import WalkForwardConfig, generate_splits

market = "IBEX"   # cambia a DJIA / SP50 / IBEX
prices = load_local(f"data/{market}.parquet")
log_ret = prices_to_log_returns(prices).dropna(how="any")

cfg = WalkForwardConfig(train_years=4, val_years=1, test_years=1, step_years=1)
splits = generate_splits(log_ret.index, cfg)

print("Nº de splits generados:", len(splits))

last = splits[-1]
print("Último fold válido:", last.fold_id)
print("train:", last.train_start.date(), "->", last.train_end.date())
print("val:  ", last.val_start.date(), "->", last.val_end.date())
print("test: ", last.test_start.date(), "->", last.test_end.date())

# Datos que quedan fuera al final
tail = log_ret.loc[last.test_end + pd.Timedelta(days=1):]
print("Filas fuera de los folds:", len(tail))
if len(tail) > 0:
    print("Rango fuera:", tail.index.min().date(), "->", tail.index.max().date())
```

```python
excluded = len(tail)
total = len(log_ret)
print(f"Fuera: {excluded}/{total} = {excluded/total:.2%}")
```
